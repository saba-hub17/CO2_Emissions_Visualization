```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("F:/Projects/co2_emissions.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
df = df.dropna()
```

```python
global_trend = df.groupby("year")["value"].sum()

plt.figure(figsize=(10,5))
plt.plot(global_trend.index, global_trend.values, marker="o")
plt.title("Global COâ‚‚ Emissions Over Time")
plt.xlabel("year")
plt.ylabel("value (metric tons)")
plt.grid(True)
plt.show()
```

```python
countries = ["Pakistan", "India", "United States", "China", "Germany"]
subset = df[df["country_name"].isin(countries)]

plt.figure(figsize=(12,6))
for c in countries:
    country_data = subset[subset["country_name"] == c]
    plt.plot(country_data["year"], country_data["value"], label=c)

plt.title("COâ‚‚ Emissions Comparison (1990â€“2022)")
plt.xlabel("year")
plt.ylabel("value (metric tons)")
plt.legend()
plt.show()
```
